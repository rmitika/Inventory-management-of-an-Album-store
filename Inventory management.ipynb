```python
ui_prod  = str(input("Enter the product_Id: "))
ui_quant = int(input("Enter the quantity: "))


print("Product: ", stock[ui_prod]['name'])
print("Price: ", stock[ui_prod]['price'])
print("Billing Amount: ", stock[ui_prod]['price'] * ui_quant)

stock[ui_prod]['quant'] = stock[ui_prod]['quant'] - ui_quant
```
